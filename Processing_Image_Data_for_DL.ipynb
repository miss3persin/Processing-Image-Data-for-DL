# Reading an image file and converting it to a numpy array

Libraries that can be used for image processing:


*   matplotlib.image
*   Pillow
*   OpenCV (cv2)



```python
# importing the image module from matplotlib

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image to matplotlib.image module

img = mpimg.imread('/content/puppy-day-e1574071869348.jpg')
```

```python
type(img)
```

```python
img.shape
# seems it's H X W X C (height by width by color channel)
```

```python
img[0, 0]
```

```python
print(img)
```

```python
# displaying the image from numpy arrray

img_plot = plt.imshow(img)
plt.show()
```

looks like the y-axis is the height and the first thing in the shape (450) while the x-axis is the width and the second thing in the shape (1096)

# Resizing an image

we need all the images to be in the same shape/dimension

Let's use the Pillow library to resize the image

```python
from PIL import Image
```

```python
img = Image.open('/content/puppy-day-e1574071869348.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized
```

```python
# saving the image

img_resized.save('resized_puppy.jpg')
```

```python
# loading the saved image

img_res = mpimg.imread('/content/resized_puppy.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
img_res.shape
```

# RGB conversion to Greyscale

we'll use OpenCV (cv2 which stands for computer vision)

```python
import cv2
```

```python
img = cv2.imread('/content/puppy-day-e1574071869348.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
